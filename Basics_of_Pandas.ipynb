```python
#Data frames are type of python dictonary

#creating a dummy data of a website
web_stats = {'Day': [1, 2, 3, 4, 5, 6],
              'Visitors': [43, 53, 34, 45, 664, 34],
              'Bounce_Rate': [65, 72, 62, 64, 54, 65]}

#converting dictonary into data-frame

df = pd.DataFrame(data= web_stats)
  
print(df.head(n = 10))   #first 10 columns
print(df.tail())         #last 5 columns
```

```python
#At present pandas is adding a index column of its own but we want index = Day column

print(df.set_index('Day'))

print(df.head())

#set_index has created a new data frame and applied the changes on it and kept the 
#original data frame unchanged; so when we print the head it returned the old data frame

#df = df.se_index('Day')  #now it will make changes on original data frame itself

#make inplace = True to make changes on data itself only
print(df.set_index('Day',inplace= True))  #modifies data inplace istelf don't create a new object
```

```python
#now the changes made by inplace= True are permanent 
#to print a specific column
print(df['Visitors'])
print(df.Visitors)
```

```python
#refrencing multiple columns
print(df[['Visitors','Bounce_Rate']])
```

```python
#Creating a list of a particular column
Visitors_List = df.Visitors.tolist()   #() are must to apply this method on dataframe

print(type(Visitors_List))
```

```python
#To convert multiple columns into numpy arrays
df_array = np.array(df[['Visitors', 'Bounce_Rate']])

print(df_array)

#Converting numpy arrays into pandas dataframe
df2 = pd.DataFrame(df_array)
print(df2)
```
